```
import pandas as pd
```

```
df_population = pd.read_csv("cc9_population_2024_percentages_for_map_corrected.csv")
df_population['district_code'] = df_population['district_code'].astype(str).str.zfill(5)
```

```
df_population
```

```
df_gps = pd.read_excel("project_chart5_gp_meds_data.xlsx")

# leading zeros
df_gps['district_code'] = df_gps['district_code'].astype(str).str.zfill(5)
```

```
df_gps
```

```
# merge for correct district_name
df_gps_merged = df_gps.merge(
    df_population[['district_code', 'district_name']],
    on='district_code',
    how='left'
)
```

```
df_gps_merged
```

```
df_gps_final = df_gps_merged[['district_code', 'district_name_y', 'gp_med']]
df_gps_final = df_gps_final.rename(columns={'district_name_y': 'district_name'})
df_gps_final.to_csv("project_chart5_gps.csv", index=False, encoding='utf-8')
print(df_gps_final.head(10))
```
