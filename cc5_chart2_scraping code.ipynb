```
import pandas as pd
```

```
# Define the URL of web-page we want to scrape from
url = "https://www.olympedia.org/statistics/medal/country"
tables_from_webpage = pd.read_html(url) # Read the tables from the webpage into a list of DataFrames
```

```
tables_from_webpage[0].head(10)
```

```
df = tables_from_webpage[0]
df.head()
```

```
print(df.columns)
```

```
df.columns = ['Country', 'Code', 'Gold', 'Silver', 'Bronze', 'Total']
```

```
df = df[['Country', 'Gold', 'Silver', 'Bronze']]
```

```
df = df.head(15)
```

```
df_tidy = pd.melt(
    df,
    id_vars=['Country'],
    value_vars=['Gold', 'Silver', 'Bronze'],
    var_name='Medal_Type',
    value_name='Count'
)
```

```
df_tidy = df_tidy.sort_values('Country').reset_index(drop=True)
df_tidy.head(10)
```

```
df_tidy.to_csv('olympic_medals_tidy.csv', index=False)
```
