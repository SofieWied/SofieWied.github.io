```python
import pandas as pd
import json
```

```python
with open('cc10_chart1_german_population_both_long_final.json', 'r') as f:
    data = json.load(f)
df = pd.DataFrame(data)
```

```python
# only 18 years old
df_18 = df[df['age'] == 18].copy()
df_18 = df_18[['year', 'population']]
```

```python
df_18
```

```python
df_18 = df_18.rename(columns={'population': 'population_age_18'})
```

```python
df_18 = df_18[(df_18['year'] >= 2024) & (df_18['year'] <= 2050)]
```

```python
df_18.to_csv('project_chart6_population_age_18.csv', index=False, encoding='utf-8')

print(df_18.head(10))
print(f"\nAnzahl Zeilen: {len(df_18)}")
```
