```
import pandas as pd
import numpy as np
```

```
df = pd.read_csv('project_chart6_personnel_input.csv')
print("Original Jahre:", df['year'].tolist())
```

```
# Interpolation
all_years = pd.DataFrame({'year': range(2024, 2050)})
df_full = all_years.merge(df, on='year', how='left')
numeric_cols = ['total_demand', 'supply_trend', 'supply_sq', 'difference_trend', 'difference_sq']
```

```
for col in numeric_cols:
    df_full[col] = df_full[col].interpolate(method='linear')

print(df_full.head(10))
print(f"\nnumber of years {len(df_full)}")
```

```
df_full.to_csv('project_chart6_personnel.csv', index=False, encoding='utf-8')
```
